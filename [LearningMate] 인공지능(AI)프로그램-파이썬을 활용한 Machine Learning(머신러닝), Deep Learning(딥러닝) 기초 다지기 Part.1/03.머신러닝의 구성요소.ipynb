###  머신러닝의 구성요소
- 데이터 준비
  - 훈련(학습)데이터 : 전체 데이터 중 모델(출력결과) 학습하기 위한 데이터
  - 검증데이터 : 전체 데이터 중 검증을 위한 데이터
  - 테스트데이터 : 현실세계에 맞닽들일 만한 데이터(현실 데이터), 최종 검증 데이터
- 모델 표현방법
  - 의사결정 트리 : 기호주의, 귀납적 추론, 철학과 심리학, 논리학에서 아이디어를 얻음
  - 신경망 기반 : 연결주의, 두뇌를 분석하고 모방하여 신경과학과 물리학에서 아이디어를 얻음
  - KNN, 서포트백터머신(SVM) : 유추주의, 유사성을 근거로 추정하면서 학습
  - 베이지안모델 : 학습이 확률 추론의 형태로 믿고 통계학에 뿌리를 둔 모델
  - 유전 알고리즘 : 진화주의, 유전학과 진화생물학에 근거를 둠
  - =======================
  - 모델 앙상블
- 모델 평가 방법
  - 에러의 제곱 : 정답 값과 모델 예측 값의 차의 제곱을 합산한 값((정답/모델예측값)^2)
  - 정확도 : 정답을 맞춘 데이터 갯수를 전체 데이터를 나눈 값
  - 우도(가늠도) : 정답에 가까울 확률
  - 정밀도와 재현률 : 정보검색에서 사용, 정밀도는 원하는 결과를 검색 결과를 나눈 값((원하는 결과값/검색결과)x100), 재현률은 원하는 사이트를 결과 사이트 총 갯수를 나눈 값((원하는 사이트 갯수/결과 사이트 총 갯수)x100)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,6,0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y1, label = "sin")
plt.plot(x,y2, linestyle = "--", label = "cos")
plt.title("sin / cos graph") # 제목

plt.xlabel("x") # x축이름
plt.ylabel("y") # y축이름
plt.legend() # 범례
plt.show()
```
